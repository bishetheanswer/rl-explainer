# Global explanations

```python
# strategy summary
chat.send_message_and_print_responses(
    """
    Based on the video, describe the overall strategy the agent has learned in one or two sentences.
    Is there a specific, advanced tactic it seems to be using?
    """
)
```

```python
# primary objective identification
chat.send_message_and_print_responses(
    """
    Based on the full gameplay clip, what was the agent's single primary objective?
    """
)
```

```python
# priority analysis
chat.send_message_and_print_responses(
    """
    What did the agent prioritize most when under pressure in this clip? e.g., points, survival,
    eliminating specific threats.
    """
)
```

```python
# playing style description
chat.send_message_and_print_responses(
    """
    Describe the agent's playing style in this clip in one or two words (e.g., "Aggressive," "Defensive,"
    "Methodical," "Erratic"). Justify your answer briefly.
    """
)
```
